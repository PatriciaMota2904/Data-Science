**Customizando Faixas de Eixos(Range)**

```
import plotly.graph_objs as go

Close = go.Scatter(
                x=df.Date,
                y=df.Close,
                name = "PETR4 High",
                line = dict(color = '#17BECF'),
                opacity = 0.8)
data = [Close]

layout = dict(
    title = "Faixa de Data Customizada",
    title_x= 0.5,
    xaxis = dict(
        range = ['2017-01-01','207-12-31'])
)
fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Manually Set Range")
```

**Rangeslider e Rangeselector**

```
import plotly.graph_objs as go

Close = go.Scatter(
    x=df.Date,
    y=df.Close,
    name = "PETR4 High",
    line = dict(color = '#17BECF'),
    opacity = 0.8)
data = [Close]

layout = dict(
    title="Série com Rangerslider e Botoes",
    title_x=0.5,
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
               dict(count=1,
                    label= '1m',
                    step='month',
                    stepmode='backward'),
               dict(count=6,
                    label= '6m',
                    step='month',
                    stepmode='backward'),     
               dict(step='all')               
          ])
            
        ),
        rangeslider=dict(
         
            visible = True
        ),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
py.iplot(fig)
```

**Trabalhando com Candlesticks**

```
trace = go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])
data = [trace]
py.iplot(data, filename='simple_candlestick')
```

**Sem rangerslider**

```
trace = go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])
layout = go.Layout(
    title = 'Preços em Candlesticks',
    title_x = 0.5,
    xaxis = dict(
        rangeslider = dict(
            visible = False
        )
        
    )
)

data = [trace]

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```
trace = go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'],
                increasing=dict(line=dict(color= '#17BECF')),
                decreasing=dict(line=dict(color= '#7F7F7F'))
                )
                
layout = go.Layout(
    title = 'Preços em Candlesticks',
    title_x = 0.5,
    xaxis = dict(
        rangeslider = dict(
            visible = False
        )
        
    )
)

data = [trace]

fig = go.Figure(data=data,layout=layout)
py.iplot(fig)
```

**Trabalhando com Médias**

```
df.head()
```

```
#Média simples de 9 dias
df['MM_9'] = df.Close.rolling(window=9).mean()

#Média simples de 17 dias
df['MM_17'] = df.Close.rolling(window=17).mean()

```

```
# Vizualizando as colunas criadas
df.head(20)
```

**Plota as médias e os preços**

```
close = go.Scatter(
                x=df.Date,
                y=df.Close,
                name = "PETR4 Close",
                line = dict(color = '#330000'),
                opacity = 0.8)
MM_9 = go.Scatter(
                x=df.Date,
                y=df['MM_9'],
                name = "Média Móvel 9 Períodos",
                line = dict(color = '#B2FF66'),
                opacity = 0.8)
MM_17 = go.Scatter(
                x=df.Date,
                y=df['MM_17'],
                name = "Média Móvel 17 Períodos",
                line = dict(color = '#FF00FF'),
                opacity = 0.8)
data = [close, MM_9, MM_17]

py.iplot(data)
```

**Trabalhando com valores menores**

```
#Cria coluna com média 3
df['MM_3'] = df.Close.rolling(window=3).mean()

#Cria coluna com média 7
df['MM_7'] = df.Close.rolling(window=7).mean()
```

```
close = go.Scatter(
                x=df.Date,
                y=df.Close,
                name = "PETR4 Close",
                line = dict(color = '#330000'),
                opacity = 0.8)
MM_3 = go.Scatter(
                x=df.Date,
                y=df['MM_3'],
                name = "Média Móvel 3 Períodos",
                line = dict(color = '#B2FF66'),
                opacity = 0.8)
MM_7 = go.Scatter(
                x=df.Date,
                y=df['MM_7'],
                name = "Média Móvel 7 Períodos",
                line = dict(color = '#FF00FF'),
                opacity = 0.8)
data = [close, MM_3, MM_7]

py.iplot(data)
```
